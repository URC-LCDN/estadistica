# Distribución Normal

La distribución normal es una distribución de probabilidad continua que sirve para modelar fenómenos naturales y sociales.

Decimos que la variable aleatoria continua X tiene una distribución normal si su función de densidad está dada por la siguiente expresión:

## $$f(x) = \frac{1}{\sqrt{2\pi σ^2}}e^\frac{(x-\mu)^2}{2σ^2} \; \; x, \mu \in R  \; , \; σ > 0$$

Decimos que $X$ se distribuye Normal, es decir, "X ∼ Normal($\mu$ , $σ^2$)"


Se utiliza ampliamente para modelar y comprender una amplia variedad de fenómenos naturales y sociales. Su utilidad radica en su capacidad para describir con precisión la distribución de datos en torno a una media, permitiendo calcular probabilidades, realizar inferencias y tomar decisiones fundamentadas en una amplia gama de disciplinas, desde la física y la biología hasta la economía y la psicología.

# Cálculo de probabilidades

Ejemplo: Un psicólogo está investigando el coeficiente intelectual (CI) en una población específica. Supongamos que los puntajes de CI en esta población siguen una distribución normal con una media de 100 y una desviación estándar de 15, lo que es común en muchos tests de CI.


# Tabla de probabilidades

```python
from scipy.stats import norm

# Parámetros
mu = 100  # media
sigma = 15  # desviación estándar
x_values = np.linspace(mu - 3*sigma, mu + 3*sigma, 11)  # rango de valores

# Generar la tabla de probabilidades
tabla_probabilidades = [(x, norm.pdf(x, mu, sigma)) for x in x_values]

# Imprimir la tabla
print("Tabla de probabilidades:")
print("x\tProbabilidad")
for x, probabilidad in tabla_probabilidades:
    print(f"{x:.2f}\t{probabilidad:.4f}")
```

1. El psicólogo está interesado en determinar qué porcentaje de individuos en esta población tiene un CI por encima de 130 y debajo de 160, que generalmente se considera como un CI muy elevado.

```python
from scipy.stats import norm

# Parámetros
mu = 100  # media
sigma = 15  # desviación estándar
x1 = 130  # límite inferior
x2 = 160  # límite superior

# Calcular la probabilidad
probabilidad_entre_x1_y_x2 = norm.cdf(x2, mu, sigma) - norm.cdf(x1, mu, sigma)
print(f"La probabilidad de que la variable esté entre {x1} y {x2} es {probabilidad_entre_x1_y_x2:.4f}")
```

# Incluso podemos calcular los percentiles

```python
from scipy.stats import norm

# Parámetros
mu = 100  # media
sigma = 15  # desviación estándar
percentil = 0.95  # percentil deseado

# Calcular el percentil
valor_percentil = norm.ppf(percentil, mu, sigma)
print(f"El percentil {percentil * 100}% de la distribución normal es {valor_percentil:.4f}")
```

# Visualización de Datos:
Para visualizar mejor los datos obtenidos de una muestra de datos que se distribuyen de manera normal con media μ y desviación estandar σ hacemos un histograma de la siguiente manera:
Histograma de la distribución Normal usando Numpy y Matplot generando 1000 valores aleatorios.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
mu = 100  # media
sigma = 15  # desviación estándar
size = 1000  # número de muestras a generar

# Generar muestras
muestras = np.random.normal(mu, sigma, size)

# Crear el histograma
plt.hist(muestras, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')

# Añadir título y etiquetas
plt.title('Histograma de muestras generadas')
plt.xlabel('Valor')
plt.ylabel('Frecuencia relativa')

# Mostrar el gráfico
plt.show()
```

# Resuelva los siguientes ejercicios:
Ingeniería y calidad del producto: Una fábrica de electrónicos produce resistencias eléctricas cuyas resistencias siguen una distribución normal con una media de 100 ohmios y una desviación estándar de 5 ohmios. La empresa establece que cualquier resistencia con una resistencia menor a 90 ohmios o mayor a 110 ohmios se considera defectuosa. Calcule la probabilidad de que una resistencia elegida al azar sea defectuosa.

```python
import scipy.stats as stats

# Definir la media y la desviación estándar
media = 100
desviacion = 5

# Calcular la probabilidad de que x esté entre 90 y 110
probabilidad = stats.norm.cdf(110, loc=media, scale=desviacion) - stats.norm.cdf(90, loc=media, scale=desviacion)

print("La probabilidad de que x esté entre 90 y 110 es:", probabilidad)
```

Economía y finanzas: Un analista financiero está estudiando el rendimiento histórico de un determinado activo financiero, como acciones o bonos. Supongamos que el rendimiento de este activo sigue una distribución normal con una media de 8% y una desviación estándar de 2%. El analista quiere calcular la probabilidad de que el rendimiento anual de este activo sea superior al 10%.

```python
# Importar la biblioteca SciPy
import scipy.stats as stats

# Definir la media y la desviación estándar
media = 8
desviacion = 2

# Calcular la probabilidad de que x sea mayor a 10
probabilidad = 1 - stats.norm.cdf(10, loc=media, scale=desviacion)

print("La probabilidad de que x sea mayor al 10%:", probabilidad)
```

Educación y evaluación: Un profesor está calificando un examen estandarizado en matemáticas en el que la puntuación de los estudiantes sigue una distribución normal con una media de 75 puntos y una desviación estándar de 10 puntos. El profesor quiere determinar qué porcentaje de estudiantes obtendrá una puntuación superior a 85 puntos en el examen.

```python
# Importar la biblioteca SciPy
import scipy.stats as stats

# Definir la media y la desviación estándar
media = 75
desviacion = 10

# Calcular la probabilidad de que x esté entre 85 y 100
probabilidad = stats.norm.cdf(100, loc=media, scale=desviacion) - stats.norm.cdf(85, loc=media, scale=desviacion)

print("La probabilidad de que x esté entre 85 y 100 es:", probabilidad)
```
